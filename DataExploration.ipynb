<H3>We would like to add a categorical value for the day of the week to be able to access it easier later on.</H3>

```python
Phil["start_time"] = pd.to_datetime(Phil["start_time"])
Phil["end_time"] = pd.to_datetime(Phil["end_time"])
```

```python
# Let's map time of day and day of week to our DF

def get_date (ts):
    return ts.date()

def get_weekday (ts):
    return ts.weekday()

def get_hour (ts):
    return ts.hour
```

```python
#Phil.reset_index(inplace=True)
Phil["Date_s"] = Phil["start_time"].apply(lambda ts: get_date (ts))
Phil["Weekday_s"]=Phil["start_time"].apply(lambda x: get_weekday (x))
Phil["Hour_s"]=Phil["start_time"].apply(lambda x: get_hour (x))
Phil["Date_e"] = Phil["end_time"].apply(lambda ts: get_date (ts))
Phil["Weekday_e"]=Phil["end_time"].apply(lambda x: get_weekday (x))
Phil["Hour_e"]=Phil["end_time"].apply(lambda x: get_hour (x))
```

<H4>We want to get rid of all data which shows a negative duration and the data with a longer duration than 24h because by Ride Indego Philadelphia policy a bike is marked as stolen and a fine is issued after the ride-time exceeds 24h. </H4>

```python
durationMin = 1 #minutes
durationMax = 1440 #minutes equals 24h

Phil.drop(Phil[Phil["duration"]<durationMin].index, inplace = True)
Phil.drop(Phil[Phil["duration"]>durationMax].index, inplace = True)

Phil
```

```python
station_coords = Phil.groupby("start_station").agg({"start_lat": "unique", "start_lon": "unique"})
```

```python
# Identify stations with multiple coordinates
stations_to_unify = station_coords[(station_coords["start_lat"].apply(len) > 1) | (station_coords["start_lon"].apply(len) > 1)].index

# Unify coordinates for stations with multiple coordinates by choosing the first pair
for station in stations_to_unify:
    latitudes = station_coords.loc[station, "start_lat"]
    longitudes = station_coords.loc[station, "start_lon"]
    unified_lat = latitudes[0]
    unified_lon = longitudes[0]
    station_coords.loc[station, "start_lat"] = [unified_lat]
    station_coords.loc[station, "start_lon"] = [unified_lon]

# Print the updated station_coords DataFrame
print(station_coords)
```

```python
# Merge station_coords with Phil based on start_station
merged_df = pd.merge(Phil, station_coords, left_on="start_station", right_index=True, how="left")

# Print the merged DataFrame
print(merged_df)

# If you want to overwrite the original DataFrame Phil with the merged DataFrame:
Phil = merged_df

# Access the unified latitude and longitude columns in the merged DataFrame
unified_latitudes = merged_df["start_lat"]
unified_longitudes = merged_df["start_lon"]
```
