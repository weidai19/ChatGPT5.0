```python
# counter how many starts at station
map2 = folium.Map(location=[39.9505, -75.1755], zoom_start = 12)

count_start = {}
for coordinate in unique_coordinates:
    count = Phil[(Phil['start_lat'] == coordinate[0]) & (Phil['start_lon'] == coordinate[1])].shape[0]
    count_start[tuple(coordinate)] = count
    
    
for coordinate, count in count_start.items():
    lat, lon, _ = coordinate
    
    if np.isnan(lat) or np.isnan(lon):
        continue 
    
    if count < 15000:
        folium.Marker(location=[lat, lon],
                 popup = f"{count}",
                  icon=folium.Icon(color="red")
                 ).add_to(map2)
        
    elif count > 15000 and count < 29999:
        folium.Marker(location=[lat, lon],
                      popup=f"{count}",
                      icon=folium.Icon(color="orange")
                     ).add_to(map2)
    else:
         folium.Marker(location=[lat, lon],
                 popup = f"{count}",
                   icon=folium.Icon(color="green")
                 ).add_to(map2)

           
map2
```

```python
# counter how many ends at station
map3 = folium.Map(location=[39.9505, -75.1755], zoom_start = 12)

count_end = {}
for coordinate in unique_coordinates_end:
    count = Phil[(Phil['end_lat'] == coordinate[0]) & (Phil['end_lon'] == coordinate[1])].shape[0]
    count_end[tuple(coordinate)] = count
    

    
for coordinate, count in count_end.items():
    lat, lon, _ = coordinate
    
    if np.isnan(lat) or np.isnan(lon):
        continue 
    
    if count < 730:
        folium.Marker(location=[lat, lon],
                 popup = f"{count}",
                  icon=folium.Icon(color="red")
                 ).add_to(map3)
        
    elif count > 100 and count < 250:
        folium.Marker(location=[lat, lon],
                      popup=f"{count}",
                      icon=folium.Icon(color="orange")
                     ).add_to(map3)
        
    elif count > 30000:
        folium.Marker(location=[lat, lon],
                      popup=f"{count}",
                      icon=folium.Icon(color="purple")
                     ).add_to(map3)
        
    else:
         folium.Marker(location=[lat, lon],
                 popup = f"{count}",
                   icon=folium.Icon(color="green")
                 ).add_to(map3)

   
map3
```

```python
# # bikes
unique_bikes = Phil["bike_id"].drop_duplicates().values.tolist()

amount_bikes = len(unique_bikes)
print (amount_bikes)


count_bikes = {}
for bike_id in unique_bikes:
    count = Phil[Phil['bike_id'] == bike_id].shape[0]
    count_bikes[bike_id] = count
    
sorted_count_bikes = dict(sorted(count_bikes.items(), key=lambda x: x[1]))
    
for bike_id, count in sorted_count_bikes.items():
    print(f"bike_id {bike_id} kommt {count} Mal vor.")
```

```python
# uses start station
unique_start_station = Phil["start_station"].drop_duplicates().values.tolist()
amount_start_station = len(unique_start_station)

count_start_station = {}
for start_station in unique_start_station:
    count = Phil[Phil["start_station"] == start_station].shape[0]
    count_start_station[start_station] = count

    
sorted_count_start_station = dict(sorted(count_start_station.items(), key=lambda x: x[1]))

for start_station, count in sorted_count_start_station.items():
    print(f" start_station {start_station} kommt {count} Mal vor")
```
